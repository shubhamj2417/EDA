**Idea**

- convert series type into dictionary

- iterate through loop

- apply thee condition get the result in the list

```python
dtype=dict(visa_df.dtypes)
dtype
```

```python
for key,value in dtype.items():
    if value in ['object']:
        print(f"{key}:{value}")
```

```python
cat_cols=[(key,value) for key,value in dtype.items() if value in ['object']]
num_cols=[(key,value) for key,value in dtype.items() if value not in ['object']]
cat_cols,num_cols
```

```python
cat_cols=[key for key,value in dtype.items() if value=='object']
num_cols=[key for key,value in dtype.items() if value!='object']
cat_cols,num_cols
```
