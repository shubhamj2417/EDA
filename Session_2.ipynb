**Create data Frame using Dictonary**

```python
names=['Avinash','Akash','Aadhya']
age=[25,30,35]
pd.DataFrame(zip(names,age),columns=['Name','Ages'])
```

```python
dict1={'Names':['Avinash','Akash','Aadhya'],
     'Ages':[25,30,35]}
dict1
```

```python
pd.DataFrame(dict1)


# keys becomes columns 
# Values becomes rows
# No need to provied columns names separately
```

```python
dict2={'Names':['Avinash'],
     'Ages':[25],
     'city':['Hyd']}
pd.DataFrame(dict2)
```
