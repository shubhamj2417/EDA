**Data Transformations**

- Generally data has 3 types of distribution 

  - Normal Distribution 
    
   - Positive skewed
    
   - Negative Skewed 
    
    
- All the maths concepts develpoed by assumption as Data follows Normal distribution

- Data skew happpens because of many reasons, in that one reason is outliers

- Once we perform the outlier analysis then we can apply transformation methods , if the data does not follow normality 

- we have some transformation methods mentioned below

   - Log transformation 
    
    - Exponential Transoformation
    
    - Reciporcal Transformation 
    
    - Square root Transformation 
    
    - Power Transformation

**Log transformation**

- Log transformation means performing logarithm operation on original data

- It is an approach to convert into Normal Distribution 

- Log means natural logarithm base=e



```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
log_data
```

```python
plt.hist(log_data,bins=20,label='Logarithm')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label='Logaritham')
plt.legend()
plt.show()
```

**Reciporcal Transformation**

- Reciprocal will fail when values are low

- for examples x=0 then 1/0 fail 

```python
x=2
1/x
```

```python
resiprocal_data=np.reciprocal(exp_data)
plt.hist(resiprocal_data,bins=20,label='Reciprocal')
plt.show()
```

**Square root transformation**

```python
sqr_data=np.sqrt(exp_data)
plt.hist(sqr_data,bins=20,label='Square Root Data')
plt.legend()
plt.show()
```

**Compare both**

```python
plt.figure(figsize=(10,5))
log_data=np.log(exp_data)
plt.subplot(1,3,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
log_data = np.log(exp_data)
plt.subplot(1,3,2).hist(log_data,bins=20,label='Logaritham')
plt.legend()
sqr_data=np.sqrt(exp_data)
plt.subplot(1,3,3).hist(sqr_data,bins=20,label='Square Root Data')
plt.legend()
plt.show()
```

**Power Transformation**

- Box cox

- ye-johnson 

- It is under sklearn preprocessing 

```python
import numpy as np
from sklearn.preprocessing import PowerTransformer
# save 
# Apply fit transform
```

```python
import numpy as np
from sklearn.preprocessing import PowerTransformer
power = PowerTransformer(method='box-cox')
exp_data1=exp_data.reshape(-1,1)
data_trans=power.fit_transform(exp_data1)
plt.hist(data_trans)
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
power = PowerTransformer(method='yeo-johnson')
exp_data1=exp_data.reshape(-1,1)
data_trans_yj=power.fit_transform(exp_data1)
plt.hist(data_trans_yj)
plt.show()
```
