**Task**

- step-1 : take prevailling wage column : **visa_df['prevailing_wage']**

    
- step -2 :calculate the mean value previlling wage: **mean= visa_df['Prevailing_wage'].mean** 
    
- step-3 : calculate the standard deviation of previlling wage: **std=visa_df['prevailing wage'].std**
    
- step-4 :calculate the Nr:step-1 - step-2 :**Nr=visa_df['preaviling_wage']-mean**

        
- step-5 divide the step4/step3
        

```python
meann=visa_df['prevailing_wage'].mean()
std=visa_df['prevailing_wage'].std()
Nr=visa_df['prevailing_wage']-meann
visa_df['prevailing_wage_z']=Nr/std
visa_df
```
