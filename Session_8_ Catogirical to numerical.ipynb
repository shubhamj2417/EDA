**One-hot Encoder**

- one hot means if one is ON and another One is OFF

- ON represents with 1, and OFF represents with 0 

- for example case study has two unique labels

  - Denied 
  
  - Crtified 
  
 - If we apply one hot encoder on case status column it will create two extra columns
 
   - case status_Denied 
   
   - Case_status_Certified
   

|Case_status |Case_status_Denied| Case_status_Certified
|-------------|----|----|
|Denied|1|0|        
|Certified|0|1|

**Adventages**

- In Machine learining we have one important propertiy is **Independent of variable 

- Machine learning models expects variables are **independent each other**

- one hot coder creating new columns , the new columns are **Independet each other**

- Independent variables means both variables have **90 Degrees phase shift**

- 90 degrees phase shift also called as **Perpendicular each other**

- perpendicular means **Othergonal each other**


- **Independent each other-90 degree - |_ - orthogonal each other - independent ecah other**

```python
**Disadvantage**

- One hot Encoder cretares new columns equal to numbers of unique labels 

- If Number of unique labels increase, new columns also increase 

- The more columns leads data is more 

- Then data is more the complexity will increase 

- complexity 
```
