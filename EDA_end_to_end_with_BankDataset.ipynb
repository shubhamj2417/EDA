**Findding the Outlier and Non outliers data**

```python
Q1=round(np.percentile(bank_df['age'], 25), 2)
Q3=round(np.percentile(bank_df['age'], 75), 2)
IQR=Q3-Q1
LB= Q1-1.5*IQR
UB= Q3+1.5*IQR
con1=bank_df['age']<LB
con2=bank_df['age']>UB
outliers_data=bank_df[con1 | con2]
len(outliers_data)
```

**Non-Outliers**

```python
Q1=round(np.percentile(bank_df['age'], 25), 2)
Q3=round(np.percentile(bank_df['age'], 75), 2)
IQR=Q3-Q1
LB= Q1-1.5*IQR
UB= Q3+1.5*IQR
con1=bank_df['age']>LB
con2=bank_df['age']<UB
non_outliers_data=bank_df[con1 & con2]
len(non_outliers_data)
```

**Histogram and Boxplot For Non-outliers data**

```python
plt.figure(figsize=(14,5))
plt.suptitle('Non outliers data')
plt.subplot(1,2,1).hist(non_outliers_data['age']) 
plt.subplot(1,2,2).boxplot(non_outliers_data['age']) 
plt.show()
```


**Comparing total dta with non outliers data**

```python
plt.figure(figsize=(10,10)) 

plt.suptitle('Non outliers data vs original data')
plt.subplot(2,2,1).hist(non_outliers_data['age']) 
plt.subplot(2,2,2).boxplot(non_outliers_data['age']) 

plt.subplot(2,2,3).hist(bank_df['age'])
plt.subplot(2,2,4).boxplot(bank_df['age']) 
plt.show()
```
